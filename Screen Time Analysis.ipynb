```python
df = pd.read_csv('Screentime - Overall Usage.csv')
```

```python
most_unlocks = df['Unlocks'].max()
print(most_unlocks)

corelated_notifications = df.loc[df['Unlocks'] == most_unlocks, 'Notifications'].iloc[0]
print('No. of times user got notification = ', corelated_notifications, ' Max number of unlocks to their phone =', most_unlocks)
```

```python
most_notifications = df['Notifications'].max()
corelated_unlocks = df.loc[df['Notifications'] == most_notifications, 'Unlocks'].iloc[0]
print('No. of times user unlocked phone = ', corelated_unlocks, ' No. of max notifications =', most_notifications)
```

```python
df['Date ']= df['Date '].astype("datetime64")
```

```python
plt.style.use("dark_background") # fivethirtyeight
df.plot(subplots=True, figsize=(12, 15))
plt.xlabel('Values')
```

```python

sorted_df = df.sort_values(by=['Total Usage'], ascending = False)
sorted_df.head()
```

```python
plt.figure(figsize=(15,8))
plt.style.use("dark_background")
sns.barplot(x='Date ',y='Total Usage', data = sorted_df.head(20))
#sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 6)
plt.ylabel('Total Usage')
plt.xlabel('Date')
plt.xticks(rotation = 60)
plt.title('Total Usage per Day')
```

```python
df1 = pd.read_csv('Screentime - App Details.csv')
```

```python
df1.head()
```

```python
plt.style.use("dark_background") # fivethirtyeight
df1.plot(subplots=True, figsize=(12, 15))
plt.xlabel('Values')
```

```python
df1.columns
```

```python
max_times = df1['Times opened'].max()
used_app = df1.loc[df1['Times opened'] == max_times, 'App'].iloc[0]

print(used_app, 'was opened max times ~', max_times, 'times')
```

```python
max_usage = df1['Usage'].max()
_app = df1.loc[df1['Usage'] == max_usage, 'App'].iloc[0]

print(_app, 'had maximum usage ~', max_usage)
```

```python
max_notification = df1['Notifications'].max()
u_app = df1.loc[df1['Notifications'] == max_notification, 'App'].iloc[0]

print(u_app, 'had max notifications ~', max_notification)
```

```python
min_notification = df1['Notifications'].min()
us_app = df1.loc[df1['Notifications'] == min_notification, 'App'].iloc[0]

print(us_app, 'had least notifications ~', min_notification)
```

```python
df1['Date']= df1['Date'].astype("datetime64")
```

```python
# Relation between Date and times an app was opened

plt.figure(figsize=(15,8))
plt.style.use("dark_background")
sns.barplot(x='Date',y='Times opened', data = df1.head(20))
#sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 6)
plt.ylabel('Times opened')
plt.xlabel('Date')
plt.xticks(rotation = 60)
plt.title('Total times apps were opened per Day')
```

```python
# Count of each app

apps = df1['App'].value_counts()
plt.figure(figsize=(15,8))
plt.bar(apps.index, apps, color=['#FFC4C4', '#F7A76C']) 
plt.xticks(rotation = 60)
plt.show()
```
